```python
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
```

```python
url='https://www.bbc.com/mundo/noticias-54347255'
page=requests.get(url).text
soup=BeautifulSoup(page,"lxml")
tabla=soup.find('table',class_='core-table gel-brevier')
print(tabla)
```

```python
country=tabla.find_all('td',attrs={'class':'c__r c__h--v'})
pais=[]
for i in range(0,len(country)):
  pais.append(country[i].get_text())
pais

#for i in tabla.find_all('tr'):
 # if i.find_all('td'.text):
 #   print(i)
```

```python
pais[:] = [paises.lstrip('\n') for paises in pais]
pais
```

```python
pais[:] = [paises.lstrip(' ') for paises in pais]
pais
```

```python
pais[:] = [paises.rstrip(' ') for paises in pais]
pais
```

```python
pais[:] = [paises.rstrip('\n') for paises in pais]
pais
```

```python
deaths=tabla.find_all('td',attrs={'class':'c__c c__c--d'})
muertes=[]
for i in range(0,len(deaths)):
  muertes.append(deaths[i].get_text())
muertes
```

```python
muertes[:] = [muert.lstrip('\n') for muert in muertes]
muertes[:] = [muert.lstrip(' ') for muert in muertes]
muertes[:] = [muert.rstrip(' ') for muert in muertes]
muertes[:] = [muert.rstrip('\n') for muert in muertes]
muertes
```

```python
death_rate=tabla.find_all('td',attrs={'class':'c__c c__c--dr'})
mortalidad=[]
for i in range(0,len(death_rate)):
  mortalidad.append(death_rate[i].get_text())
mortalidad
```

```python
mortalidad[:] = [mort.lstrip('\n') for mort in mortalidad]
mortalidad[:] = [mort.lstrip(' ') for mort in mortalidad]
mortalidad[:] = [mort.rstrip(' ') for mort in mortalidad]
mortalidad[:] = [mort.rstrip('\n') for mort in mortalidad]
mortalidad
```

```python
cases=tabla.find_all('td',attrs={'class':'c__c c__c--t'})
total_de_casos=[]
for i in range(0,len(cases)):
  total_de_casos.append(cases[i].get_text())
total_de_casos
```

```python
total_de_casos[:] = [casos.lstrip('\n') for casos in total_de_casos]
total_de_casos[:] = [casos.lstrip(' ') for casos in total_de_casos]
total_de_casos[:] = [casos.rstrip(' ') for casos in total_de_casos]
total_de_casos[:] = [casos.rstrip('\n') for casos in total_de_casos]
total_de_casos
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df['PAIS']=pais
df['MUERTES']=muertes
df['MORTALIDAD']=mortalidad
df['CASOS']=total_de_casos
df
```

```python
df.to_csv(r'E:\reviews.csv',index=True)
```
